```python
import pandas as pd

path_df1 = "C:/Users/osanc/Desktop/IronHack/Sem_3/Proyecto/pip.csv"
path_df2 = "C:/Users/osanc/Desktop/IronHack/Sem_3/Proyecto/completion-rate-of-upper-secondary-education-sdg.csv"
path_df3 = "C:/Users/osanc/Desktop/IronHack/Sem_3/Proyecto/expected-years-of-schooling-vs-share-in-extreme-poverty.csv"
path_df5 = "C:/Users/osanc/Desktop/IronHack/Sem_3/Proyecto/total-government-expenditure-on-education-gdp.csv"
path_df6 = "C:/Users/osanc/Desktop/IronHack/Sem_3/Proyecto/country-coord.csv"

df1 = pd.read_csv(path_df1)
df3 = pd.read_csv(path_df3)
df5 = pd.read_csv(path_df5)
df6 = pd.read_csv(path_df6)
```

```python
df1 = df1[['region_name', 'region_code', 'country_name', 'country_code','reporting_year','gini','poverty_line', 'headcount','poverty_gap','reporting_pop', 'reporting_gdp']]
```

```python
df1["codigo"] = df1['country_code'] + "_" + df1['reporting_year'].astype(str)
```

```python
df3.head(3)
```

```python
df3["codigo"] = df3['Code']+"_"+df3['Year'].astype(str)
```

```python
df3 = df3[["codigo","Expected years of schooling"]]
```

```python
df3.rename(columns={"Expected years of schooling":"expected_years_school"}, inplace=True)
```

```python
df3.head()
```

```python
df5.head(3)
```

```python
df5["codigo"] = df5['Code']+"_"+df5['Year'].astype(str)
```

```python
df5 = df5[["codigo","Historical and more recent expenditure estimates"]]
```

```python
df5.rename(columns={"Historical and more recent expenditure estimates":"spend_public_education"}, inplace=True)
```

```python
df5.head(3)
```

```python
df6.head(3)
```

```python
df6.rename(columns={"Alpha-3 code":"country_code","Latitude (average)":"Latitude","Longitude (average)":"Longitude"}, inplace=True)
```

```python
df6 = df6[["country_code","Latitude","Longitude"]]
```

```python
df_merged = pd.merge(df1, df3, on='codigo', how='left')
df_merged = pd.merge(df_merged, df5, on='codigo', how='left') 
df_merged = pd.merge(df_merged, df6, on='country_code', how='left')
```

```python
df_merged.head()
```

```python
df_merged.info()
```

```python
df_merged = df_merged[(df_merged["reporting_year"]>=2000) & (df_merged["reporting_year"]<=2019)]
```

```python
df_merged.isnull().sum()
```

```python
# NÃºmero de columnas en el DataFrame
num_columns = len(df_merged.columns)

```

```python
df_merged = df_merged.dropna(thresh=num_columns - 3)
```

```python
df_merged.isnull().sum()
```

```python
df_merged['expected_years_school'] = df_merged.groupby('country_name')['expected_years_school'].transform(lambda x: x.fillna(x.mean()))
```

```python
df_merged['spend_public_education'] = df_merged.groupby('country_name')['spend_public_education'].transform(lambda x: x.fillna(x.mean()))
```

```python
df_merged.isnull().sum()
```

```python
df_merged = df_merged.dropna(subset=["expected_years_school", "spend_public_education"])
```

```python

df_merged.groupby(["region_name","country_name"]).agg({
    'gini': 'mean',
    'headcount': 'mean',
    "expected_years_school":"mean",
    "spend_public_education":"mean"})
```

```python
# Usar pivot_table para obtener el formato deseado
df_filtered = df_merged.loc[df_merged['country_name']=="Peru"]

df_pivot = df_filtered.melt(id_vars=['country_name', 'reporting_year'], 
                            value_vars=['gini', 'Expected years of schooling'], 
                            var_name='indicator')

# Crear la tabla dinÃ¡mica
result = df_pivot.pivot_table(index='indicator', columns='reporting_year', values='value')
```
